```python
def crawl_btc_price_api():
  res = requests.get('https://www.binance.com/en/trade/BTC_USDT?_from=markets&theme=dark&type=spot')
  btc_price = float(re.findall(r'"close":"\d+.\d+"', res.text)[0].split(":")[1].split('"')[1])
  now = datetime.datetime.now()
  dt_string = now.strftime("%Y-%m-%d %H:%M:%S")
  btc_info = {'Symbol':'BTC','Price':btc_price,'Updated_At':dt_string}
  res.close()
  return btc_info
```

```python
pbi_url = 'https://api.powerbi.com/beta/6ac2ad06-692c-4663-b7af-a9ff2a866d0c/datasets/5cd6eebd-0139-4b27-8226-e8dd3a84f7c8/rows?experience=power-bi&key=RR3AJRvq1do%2B7mUgO8jfZ1hXRReC%2BywbOeODpNNXvyJ%2Bvcyle%2FLnGK%2BQ11DtT5VVtbCg1Lw5cEQFu%2B6QoIoFuA%3D%3D'
headers = {'Content-Type':'application/json'}
while True :
    get_btc_price = crawl_btc_price_api()
    get_btc_price = [get_btc_price]
    response = requests.request(method = "POST",url=pbi_url,headers=headers,data = json.dumps(get_btc_price))
    response.close()
    print(get_btc_price)
    time.sleep(120)
```
